# Files and Folders.

```python
import os, pathlib

print(os.getcwd())
print(pathlib.Path.cwd())

print()
filePath = pathlib.Path(r"C:\Windows\System32\calc.exe")
print(filePath.name); print(filePath.stem); print(filePath.suffix)
print(filePath.parent)

print()
#Checking for file existence
print(filePath.is_file())
print(filePath.exists())

print()
#Now checking the home directory
print(pathlib.Path.home())
```
