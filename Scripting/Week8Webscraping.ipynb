```python
import requests
from bs4 import BeautifulSoup

url = "https://www.meteoblue.com/en/weather/webcams/madrid_spain_3117735"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
images = soup.find_all("a", {'class' : 'webcam-img'})
#print(images)

pictureLinks = []
for tag in images:
    pictureLinks.append(tag.img["src"])
print(pictureLinks)

index = 1
for newURL in pictureLinks:
    filename = "webcams/original" + str(index) + ".jpg"
    r = requests.get(newURL, allow_redirects=True)
    with open(filename, 'wb') as file:
        file.write(r.content)
    index += 1
    

    
```

```python
import datetime as dt
print(dt.datetime.now())
date = dt.datetime(2000, 5, 10, 0, 0, 0)
print(date)

# date from a number of seconds
print(dt.datetime.fromtimestamp(1770000000))
print(dt.datetime.fromtimestamp(time.time()).year + 1)

# FORMATTING  --> strftime() method
print()
print(date)
print(date.strftime("%d %B %y"))
print(date.strftime("%d %B %Y"))
print(date.strftime("%D"))

date += dt.timedelta(days = 1000)
print(date.strftime("%d %b %Y"))
```

## Multithreading
Python default program always runs in single threaded manner, so it only executes one thing at a time. A multithreaded program can execute multiple things at the same time.

```python
import time

def sleep():
    time.sleep(5)
    print("HELLOOOO I AM AWAKE")
    
def multi(number):
    for i in range(0, 10):
        print(f'{i} * {number} = {i*number}')
        
```

```python
import time, threading

def sleep():
    time.sleep(5)
    print("HELLOOOO I AM AWAKE")
    
def multi(number):
    for i in range(0, 10):
        print(f'{i} * {number} = {i*number}')
        
# MULTITHREADING
sleeper = threading.Thread(target=sleep)
muliplicator = threading.Thread(target=multi(8))

print("START")
sleeper.start()
muliplicator.start()
print("STOP")
```

```python
import time, threading

def function1():
    for i in range(1, 11):
        print(i, end=" ")
        time.sleep(0.5)

def function2():
    for c in "ABCDEFGHIJ":
        print(c, end=" ")
        time.sleep(0.51)

thread1 = threading.Thread(target=function1)
thread2 = threading.Thread(target=function2)

print("START")
thread1.start()
thread2.start()
thread1.join()
thread2.join()  
print("STOP")

# if you encounter CONCURRENCY, add 0.01 seconds to one of the timers.
```
