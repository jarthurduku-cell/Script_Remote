## Input and output
Conversion between datatypes
- int --> whole numbers
- float --> decimal value
- str --> string of text

```python
#find the sign of x
x = input()
print(x)
print(type(x))

int_x = int(x)
print(int_x)
print(type(int_x))

float_x = float(x)
print(float_x)
print(type(float_x))

backToString = str(float_x)
print(backToString)
print(type(backToString))
```

## Combining outputs -->  concatenation
- '+' --> combining 2 strings or adding 2 values toogether if thery can be calculated with.
- ',' --> split between 2 datatypes to be able to combine them into one piece of output/text
- f"" --> allows us to add variables in between {} into a string of text

```python
number = 6
print("The number is:",number) # careful with this one, it will add a space between the text and the number
print(f"The number is: {number}")
print("The number is: " + str(number)) # using concatenation does not automatically add a space

print(number + number)
#print(str(number) + number)
print(str(number) + str(number))
print(str(number + number))
#print("6" + number)
```

### Some exercises on if-statements
- Write a program that checks if you're eligible to vote, you have to be 18 or older and your age should be calculated based upon the current year (using datetime module) --> if eligible: "ELIGIBLE", otherwise: "NOT ELIGIBLE"

- Write up 2 ways of how you can check if a string is empty with an if-statement.

```python
from datetime import datetime as dt

yob= int(input())

current_year = dt.now().year

age = current_year - yob

if age >= 18:
    print("ELIGIBLE")
else:
    print("NOT ELIGIBLE") 
```

```python
if(input() == ""):
    print("Empty")
else:
    print("Full")
    
    
if(len(input()) == 0):
    print("Empty")
else:
    print("Full")
    

s = input()
if not s:
    print("Empty")
else:
    print("Full")    
    
```

### Exercise for loops:
FIZZBUZZ
Write a program that prints out the numbers from 1 --> 100 and think of the following
"Fizz" instead of the number, if divisible by 3 (only 3)
"Buzz" instead of the number, if it is divisible by 5 (only 5)
"FizzBuzz" instead of the number if it is divisible by both 3 and 5

```python
for i in range(1,101):
    if i % 3 == 0:
        print("Fizz", end=" ")
    elif i % 5 == 0:
        print("Buzz", end=" ")
    elif i % 5 == 0 and i % 3 == 0:
        print("FizzBuzz", end=" ")
    else:
        print(i, end=" ")
```
