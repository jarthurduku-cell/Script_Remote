## Last exercise on regex
Write a python function --> validate_email() that takes in an email address and returns a boolean that tells the user whether the email is valid or not.
- Emailaddress has to contain a @
- First part of the email ( before @ ) can only include letters, digits, hyphens ( - ), underscores ( _ ) and dots. Has to be at least 1 char.
- part after the @ --> has to be at least 2 chars, and only letters, dots . and hyphens ( - )
- cannot end in a dot.


```python
import re
def validate_email():
    email = input("Enter your email: ")
    
    regex = r"[A-z0-9\-_\.]+@[A-z\-\.]{2,}"
    list = re.findall(regex, email)
    
    for item in list:
        if item[-1] == ".":
            return "False"
        else:
           return "True"
   
    
validate_email()
    
```

```python
import re
def validate_email(email):
    if "@" not in email:
        return False
    
    parts = email.split("@")
    if len(parts) != 2:
        return False
    
    regex = r"^[\w\.\-_]+$"
    if not re.match(regex, parts[0]):
        return False
    
    regex = r"^[A-z\-\.]{2,}$"
    if re.match(regex, parts[0]) is None:
        return False
    
    if email[-1] == ".":
        return False
    
    
    return True

print(validate_email("anthony.coppens@thoma-smore.be"))
print(validate_email("x!,anthony.coppens!:;@thomas-more.be"))

```
