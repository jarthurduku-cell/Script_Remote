#### Sequence of Hailstone
### If n is 1 then the sequence ends.
### If n is even then the next n of the sequence = n/2
### If n is odd then the next n of the sequence = (3 * n) + 1

```python
n = int(input("Enter a number:"))
print("n is",n)

while not(n == 1):
    if n % 2 == 0:
        n//=2
    else:
        n = (3*n) + 1
    print(n)
```

## Functions

```python
def sign(x):
    if x > 0:
        return "Positive"
    elif x < 0:
        return "Negative"
    else: 
        return "Zero"

x = int(input("Enter a number:"))  
print(x,"is",sign(x))

print()
## Trying with flow control statements like for loop
for z in range(-1,2):
    print(z,"is",sign(z))
```

Python list  ------> C# array( mixed data types)
Python tuple ------> C# string array (uniform data types)
Python set   ------> C# list
Py dictionary -----> C# dictionary


```python
#  Tuples are immutable lists
# Tuples use normal parentheses.
inties = (1, 2, 3)
strings = ("foo", "bar", "baz")
chars = ('a', 'b', 'c')
boolies = (True, False, True)
print(inties); print(strings); print(chars); print(boolies)
print()

#Sets are unordered collections of unique elements.
# They use curly braces.
setNum = {1, 2, 3, 4, 5}
setNum.add(8)

setStr = {"foo", "bar", "baz"}
setStr.add("love")

setChar = {'a', 'b', 'c'}
setChar.add('d')

setBool = {True, False}
setBool.add("True") # THe bool should be added as a string.Beccause duplicates are not allowed in sets.

print(setNum); print(setStr); print(setChar); print(setBool)
```

```python
# Dictionaries
# Dictionaries are collections of key-value pairs.
# They use curly braces and colons to separate keys and values.
dictNum = {1: "one", 2: "two", 3: "three"}
dictNum[4] = "four"
dictStr = {"foo": 1, "bar": 2, "baz": 3}
dictStr["Love"] = 7
dictChar = {'a': "alpha", 'b': "beta", 'c': "gamma"}
dictChar['d'] = "delta"
print(dictNum); print(dictStr); print(dictChar)
```
