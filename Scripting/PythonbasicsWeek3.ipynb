## Slicing arrays
[start : end : step]
- start: number (index), included
- end: number (index), NOT included
- step: the size of a jump between values



```python
import numpy as np

array = np.array([6, 8, 2, 9, 0, 1, 3, 5, 7, 4])
print(array)

# I want the first 4 numbers of the array
print(array[0:4:1])
print(array[0:4])
print(array[:4:])

# I want the 8 2 9 0
print(array[1:5:])

# I want the 8 9 
print(array[1:5:2])
print(array[1:4:2])

print(array[::]) # gives everything

#reverse
print(array[::-1])
print(array[-1::])

#slice the last 4, but don't use the positive index
print(array[-4::])
print(array[-1:0:-1]) #[4 7 5 3 1 0 9 2 8]
print(array[0:10:-1]) # prints nothing
```

### Exercise on slicing
we want to generate an array of the temperature measurements EVERY hour throughout a day.
(24 measurements) 
generate random temps: between 0 and 25 (use random ints)
--> use slicing to isolate periods and analyse them
- earlyBirdFeedsTheWorm: all temps first 6 hours of the day (includes the temp at 00:00 --> 06:00)
- everyoneAwakeBEFORENOON: 09:00 --> 13:00 (again the temps here)
- siesta: 14:00 - 16:00
- lateNightOwls: because they live in a different time, reverse the slice (LAST 4 MEARSURMENTS)
- trending: every 5 hours, starting at 00:00, 05:00, 10:00....


```python
import numpy as np
import random

#temps = [random.randint(0,26) for x in range(24)]
#print(temps)

measurments  = np.random.randint(0,26,20)
print(measurments)

earlyBirdFeedsTheWorm = measurments[:7:]
print(earlyBirdFeedsTheWorm)

everyoneAwakeBEFORENOON = measurments[9:14:]
print(everyoneAwakeBEFORENOON)

siesta = measurments[14:17:]
print(siesta)

lateNightOwls = measurments[-1:-5:-1]
print(lateNightOwls)

trending = measurments[::5]
print(trending)



```

# Working with arrays

```python
import numpy as np

a = np.arange(0,10)
b = np.arange(9, -1, -1)
print(a,b)

print(a + b)
print(a - b)
print(a * b)

print(np.where(a % 2 == 0, a + 5, a))
print(np.where(a % 2 == 0))

array = np.where(a % 2 == 0)
print(array[0]) # the regular np.where with only a condition and no True or False, returms a tuple with the indexes 
# of the values where the condition is met.

print(np.where(b > 5))
meetingCondition = np.where(b > 5)[0]
print(meetingCondition)

b = np.reshape(b, (2,5))
print(b)
print()
print(b.sum(axis = 0)) # the sum of all columns each counted up together
print(b.sum(axis = 1)) # sum of each row
print(b.cumsum(axis = 1))

print()
print(a.sum())
print(a.min())
print(a.max())

print(np.sqrt(a)) # square root of each number in a
print(np.sqrt(a.max()))
```
