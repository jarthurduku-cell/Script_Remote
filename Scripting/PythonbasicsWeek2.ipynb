```python
def CountVowelPARNoRe(word):
    counter = 0
    for c in word.lower():
        if c == 'a' or c == 'e' or c == 'i' or c == 'o' or c == 'u':
            counter += 1
    print(word, "has", counter, "vowels")

CountVowelPARNoRe(input())
```

```python
def CountVowelNOPARYESreturn():
    counter = 0
    word = input("Enter a word: ").lower()
    for c in word: # ord() --> transforms a character into its ascii/numerical value
        if ord(c) == 97 or ord(c) == 101 or ord(c) == 105 or ord(c) == 111 or ord(c) == 117:
            counter += 1
    return counter

print(CountVowelNOPARYESreturn(), "vowel(s)")
```

```python
def CountVowelPR(word):
    counter = 0
    
    for c in word:
        if c in "aeiou":
            counter += 1
    return counter

word = input("Enter a word: ").lower()
print(word, "has", CountVowelPR(word), "vowels")
```

### Dictionary { key : value}
-> unordered, no indexes, they are changeable, but indexed by key
-> contains key-value pairs split with a : and created with { }
Can contain any datatype as both key and value, only part that truely needs to be is that the keys must be hashable(immutable)

```python
dictionary = { "a" : "apple", "b" : "banana", "c" : "coconut" , "d" : "dragonfruit" }
print(dictionary)

dictionary["c"] = "carrot"
print(dictionary)

dictionary["e"] = "peach"
print(dictionary)

dictionaryList = { "a" : ["apple","apricot"], "b" : ["banana"] , "c" : ["coconut", "carrot", "cucumber"] }
print(dictionaryList)

dictionaryList["a"].append("a other fruit")
print(dictionaryList)

for key in dictionaryList:
    print(key, "->")
    for value in dictionaryList[key]:
        print(value)
```
