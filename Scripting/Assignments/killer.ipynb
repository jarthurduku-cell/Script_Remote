```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import json
import os
import unicodedata


def top_killers(n, sort_by, location):
    if " " in location:
        location = location.replace(" ", "-")
    if location != "world":
        url = f"https://killer.cloud/serial-killers/by/country/{location.lower()}"
        headers = {'User-Agent' : 'Anthony'}
        response = requests.get(url, headers=headers)
        #print(response.status_code)
        #html = response.text    
        soup = BeautifulSoup(response.content, 'html.parser')

        naamen = soup.find_all("h3", class_=lambda c: c and "pad10t" in c and "mar5t" in c and "font-hel" in c)
        victims = soup.find_all("h4", class_=lambda c: c and "mar0" in c and "pad10t" in c and "font-hel" in c)
        act_years = soup.find_all("h4", class_=lambda c: c and "mar0" in c and "pad15tb" in c and "font-hel" in c)
        links = soup.find_all("a", {'class' : 'bl text-left mar5tb'})

        finalList = []
        for i in range(min(len(naamen), len(victims), len(act_years), len(links))):
            diction = {'name' : naamen[i].get_text(),'victims': int(victims[i].get_text()),'years' : int(act_years[i].get_text()),'url' :  links[i]['href']}
            finalList.append(diction)

        p = 2    
        while len(finalList) < n:
            url =  f"https://killer.cloud/serial-killers/by/country/{location}?page={p}"
            p += 1
            headers = {'User-Agent' : 'Anthony'}
            response = requests.get(url, headers=headers)   
            soup = BeautifulSoup(response.content, 'html.parser')

            naamen = soup.find_all("h3", class_=lambda c: c and "pad10t" in c and "mar5t" in c and "font-hel" in c)
            victims = soup.find_all("h4", class_=lambda c: c and "mar0" in c and "pad10t" in c and "font-hel" in c)
            act_years = soup.find_all("h4", class_=lambda c: c and "mar0" in c and "pad15tb" in c and "font-hel" in c)
            links = soup.find_all("a", {'class' : 'bl text-left mar5tb'})

            if not naamen:
                break
            for i in range(min(len(naamen), len(victims), len(act_years), len(links))):
                diction = {'name' : naamen[i].get_text(),'victims': int(victims[i].get_text()),'years' : int(act_years[i].get_text()),'url' :  links[i]['href']}
                finalList.append(diction)
   
        if n < len(finalList):
            return finalList[:n]
        else:
            return finalList
        
    elif location == "world": 
        if sort_by == "confirmed-kills":
            url = f"https://killer.cloud/serial-killers/by/{sort_by}?page=1"
            headers = {'User-Agent' : 'Anthony'}
            response = requests.get(url, headers=headers)
            #print(response.status_code)
            #html = response.text    
            soup = BeautifulSoup(response.content, 'html.parser')

            naamen = soup.find_all("h3", class_=lambda c: c and "pad10t" in c and "mar5t" in c and "font-hel" in c)
            victims = soup.find_all("h4", class_=lambda c: c and "mar0" in c and "pad10t" in c and "font-hel" in c)
            act_years = soup.find_all("h4", class_=lambda c: c and "mar0" in c and "pad15tb" in c and "font-hel" in c)
            links = soup.find_all("a", {'class' : 'bl text-left mar5tb'})

            finale = []
            for i in range(min(len(naamen), len(victims), len(act_years), len(links))):
                diction = {'name' : naamen[i].get_text(),'victims': int(victims[i].get_text()),'years' : int(act_years[i].get_text()),'url' :  links[i]['href']}
                finale.append(diction)
            p = 2
            #print(finale, "just the list")
            
                
            while len(finale) < n:
                url = f"https://killer.cloud/serial-killers/by/{sort_by}?page={p}"
                p += 1
                headers = {'User-Agent' : 'Anthony'}
                response = requests.get(url, headers=headers)
                #print(response.status_code)
                #html = response.text    
                soup = BeautifulSoup(response.content, 'html.parser')

                naamen = soup.find_all("h3", class_=lambda c: c and "pad10t" in c and "mar5t" in c and "font-hel" in c)
                victims = soup.find_all("h4", class_=lambda c: c and "mar0" in c and "pad10t" in c and "font-hel" in c)
                act_years = soup.find_all("h4", class_=lambda c: c and "mar0" in c and "pad15tb" in c and "font-hel" in c)
                links = soup.find_all("a", {'class' : 'bl text-left mar5tb'})

                for i in range(min(len(naamen), len(victims), len(act_years), len(links))):
                    diction = {'name' : naamen[i].get_text(),'victims': int(victims[i].get_text()),'years' : int(act_years[i].get_text()),'url' :  links[i]['href']}
                    finale.append(diction)
    
            if n < len(finale):
                return finale[:n]
            else:
                return finale

        elif sort_by == "years-active":

            url = f"https://killer.cloud/serial-killers/by/{sort_by}"
            headers = {'User-Agent' : 'Anthony'}
            response = requests.get(url, headers=headers)
            #print(response.status_code)
            #html = response.text    
            soup = BeautifulSoup(response.content, 'html.parser')

            naamen = soup.find_all("h3", class_=lambda c: c and "pad10t" in c and "mar5t" in c and "font-hel" in c)
            victims = soup.find_all("h4", class_=lambda c: c and "mar0" in c and "pad10t" in c and "font-hel" in c)
            act_years = soup.find_all("h4", class_=lambda c: c and "mar0" in c and "pad15tb" in c and "font-hel" in c)
            links = soup.find_all("a", {'class' : 'bl text-left mar5tb'})


            finale = []
            for i in range(min(len(naamen), len(victims), len(act_years), len(links))):
                diction = {'name' : naamen[i].get_text(),'victims': int(victims[i].get_text()),'years' : int(act_years[i].get_text()),'url' :  links[i]['href']}
                finale.append(diction)
            p = 2

            while len(finale) < n:
                url = f"https://killer.cloud/serial-killers/by/{sort_by}?page={p}"
                p += 1
                headers = {'User-Agent' : 'Anthony'}
                response = requests.get(url, headers=headers)
                #print(response.status_code)
                #html = response.text    
                soup = BeautifulSoup(response.content, 'html.parser')

                naamen = soup.find_all("h3", class_=lambda c: c and "pad10t" in c and "mar5t" in c and "font-hel" in c)
                victims = soup.find_all("h4", class_=lambda c: c and "mar0" in c and "pad10t" in c and "font-hel" in c)
                act_years = soup.find_all("h4", class_=lambda c: c and "mar0" in c and "pad15tb" in c and "font-hel" in c)
                links = soup.find_all("a", {'class' : 'bl text-left mar5tb'})

                for i in range(min(len(naamen), len(victims), len(act_years), len(links))):
                    diction = {'name' : naamen[i].get_text(),'victims': int(victims[i].get_text()),'years' : int(act_years[i].get_text()),'url' :  links[i]['href']}
                    finale.append(diction)
   
            if n < len(finale):
                return finale[:n] 
            else:
                return finale


def analyse(killers):
 
    with open("mafia.csv", "w", encoding="utf-8") as mafia:
        mafia.write("name,victims,years,url\n")

        for x in killers:
            line = f'{x["name"].replace(","," ")},{int(x["victims"])},{int(x["years"])},{x["url"]}\n'
            mafia.write(line)
    boss = pd.read_csv("mafia.csv", sep = ",")

    data = boss.describe()
    count = round(data.loc["count","victims"])   
    avg_vic = round(data.loc["mean","victims"],1)  
    max_vict = round(data.loc["max","victims"])

    k_max_vict = ""
    corr_year = boss[boss["victims"] == max_vict].iloc[0,2]
    for x in killers:
        if x["victims"] == max_vict and x["years"] == corr_year:
            k_max_vict = x["name"]
    
    most_active = round(data.loc["max","years"])
    k_most_active = ""
    corr_vic = boss[boss["years"] == most_active].iloc[0,1]
    for x in killers:
        if x["years"] == most_active and x["victims"] == corr_vic:
            k_most_active = x["name"]

    
    
    return f"--- Analysis ---\nKillers analysed: {count}\nAverage victims: {avg_vic}\nMost victims: {k_max_vict} ({max_vict})\nLongest active: {k_most_active} ({most_active} years)"

def profile(name):
    nameCode = name.replace(" ", "+")
    #https://killer.cloud/search?term=Jeffrey+Dahmer
    url = f"https://killer.cloud/search?term={nameCode}"
    headers = {'User-Agent' : 'Anthony'}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    
    link = soup.find("a", {'class' : 'bl text-left mar5tb'})

    
    
    profile_url = link['href']
    headers = {'User-Agent' : 'Anthony'}
    p_response = requests.get(profile_url, headers=headers)
    p_soup = BeautifulSoup(p_response.content, 'html.parser')

    
    naam = p_soup.find("span", {'class' : 'wikidata'})
    name = naam.get_text()
    
    list_nick = []
    c = 1
    tr = p_soup.find_all("tr")
    for x in tr:
        list_nick.append(x.get_text())
        c += 1
   
    ch = (z for z in list_nick if "Sentence" in z)
    val = list(ch)

        
    sen = list_nick.index(val[0])
    new_list = list_nick[:sen + 1]
    fin_list = []
    for x in new_list:
        if ":" not in x:
            new_list.remove(x)
     
    for x in new_list:
        fin_list.append(x.split(":"))       
        
    dict = {}
    for key, value in fin_list:
        dict[key.strip()] = value.strip()
 
     
    nickname = dict.get("Nickname")
    victims = dict.get("Victims") 
    years_active = dict.get("Years Active")
    countries = dict.get("Active Countries")
    gender = dict.get("Gender")
    nationality = dict.get("Citizenship")
    iq = dict.get("IQ")
    arrested = dict.get("Arrested")
    convicted = dict.get("Convicted")
    sentence = dict.get("Sentence")

        
    final_dict = { "name" : name, "nickname" : nickname, "victims" : victims, "years_active" : years_active.replace(" ", ""), "countries": countries, "gender" : gender, "citizenship" : nationality, "iq" : iq, "arrested" : arrested, "convicted" : convicted, "sentence" : sentence }
    return final_dict  


def summary(name):
    
    name = name.replace(" ", "_")
    
    url = f"https://en.wikipedia.org/api/rest_v1/page/summary/{name}"
    headers = {'User-Agent' : 'Anthony'}
    response = requests.get(url , headers=headers)

    if response.status_code == 200:
        data = json.loads(response.text)
        summary = data["extract"]
        return summary
    else:
        return "No Wikipedia entry found."

def profilePicture(name):

    naam = name
    #name = unicodedata.normalize('NFKD', name)
    name = ''.join(c for c in name if not unicodedata.combining(c))
    sex = profile(name)["gender"]
    name = name.split(" ")
    lastName = name[-1]
    initial = name[-1][0]
    nameLink = name[-1] + "-" + "-".join(name[:-1])
    
    
    sex = "Male"
    url = rf"https://murderpedia.org/{sex.lower()}.{initial}/{initial.lower()}/{nameLink.lower()}.htm"
    headers = {'User-Agent' : 'Anthony'}
    response = requests.get(url, headers=headers)
    code = response.status_code
    
    num = 1
    if code != 200:
        while code != 200:
                url = rf"https://murderpedia.org/{sex.lower()}.{initial}/{initial.lower()}{num}/{nameLink.lower()}.htm"
                num += 1 
                headers = {'User-Agent' : 'Anthony'}
                response = requests.get(url, headers=headers)
                code = response.status_code

                if num == 5:
                    break
    if code != 200:
        sex = "Female"
        url = rf"https://murderpedia.org/{sex.lower()}.{initial}/{initial.lower()}/{nameLink.lower()}.htm"
        print(url)
        headers = {'User-Agent' : 'Anthony'}
        response = requests.get(url, headers=headers)
        code = response.status_code
        
        if code != 200:
            while code != 200:
                url = rf"https://murderpedia.org/{sex.lower()}.{initial}/{initial.lower()}{num}/{nameLink.lower()}.htm"
                num += 1 
                headers = {'User-Agent' : 'Anthony'}
                response = requests.get(url, headers=headers)
                code = response.status_code

                if num == 5:
                    break
    
    soup = BeautifulSoup(response.content, 'html.parser')
    image = soup.find_all("img")
    image_url = ""
    for img in image:
        
        imageLink = img["src"]
        if imageLink.startswith("../images"):
            image_url = imageLink
            break

    part = f"https://murderpedia.org/{sex.lower()}.{initial}"
    
    image_url = image_url.replace("..", part)

    filename = f"""{naam.replace(" ","_").lower()}""" + ".jpg"
    
    r = requests.get(image_url, allow_redirects=True)
    with open(filename, 'wb') as file:
        file.write(r.content)
    return filename
    
    
    
   

def main():
    country = input("Enter a country (or 'world'):")
    killerNumb = ""
    sort = ""
    top = ""
    if country != "world":
        killerNumb = int(input(" How many killers? "))
        
    else: 
        sort = input(" Sort by (confirmed-kills / years-active):") 
        killerNumb = int(input(" How many killers? "))
        
        if sort == "years_active":
            sort = sort.replace("_","-")

    top = top_killers(killerNumb, sort, country)   
    print()
    count = 1
    for dic in top:
        print(f"{count}. {dic['name']} — {dic['victims']} victims — {dic['years']} years active")
        count += 1
    
    analysis = analyse(top)
    print(f"\n{analysis}\n")

    killerName = input("Enter a killer name to investigate: \n")
    profiel = profile(killerName)
    print(f"--- {profiel['name']} ---")
    print(f"Nickname:	{profiel['nickname']}")
    print(f"Victims:	{profiel['victims']}")
    print(f"Years active:	{profiel['years_active']}")
    print(f"Countries:	{profiel['countries']}")
    print(f"Gender:		{profiel['gender']}")
    print(f"IQ:		{profiel['iq']}")
    print(f"Arrested:	{profiel['arrested']}")
    print(f"Convicted:	{profiel['convicted']}")
    print(f"Sentence:	{profiel['sentence']}")

    summarize = summary(killerName)
    print("\n--- Wikipedia ---")
    print(summarize)
    
    picture = profilePicture(killerName)
    if picture:
        print(f"\nProfile picture saved as {picture}")
    else:
        print(f"""\nProfile picture saved as {killerName.replace(" ","_").lower()}.jpg""")



if __name__ == "__main__":
    main()
```
