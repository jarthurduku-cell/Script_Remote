```python
import json
import requests
from bs4 import BeautifulSoup
import time
import datetime as dt
import threading
import pandas as pd
from dateutil import parser


def zodiac(day, month):
    if month == 1 and day >= 20 and day <= 31 or month == 2 and day >= 1 and day <= 18:
        return f"Zodiac sign: Aquarius"
    elif month == 2 and day >= 19 and day <= 29 or month == 3 and day >= 1 and day <= 20:
        return f"Zodiac sign: Pisces"
    elif month == 3 and day >= 21 and day <= 31 or month == 4 and day >= 1 and day <= 19:
        return f"Zodiac sign: Aries"
    elif month == 4 and day >= 20 and day <= 30 or month == 5 and day >= 1 and day <= 20:
        return f"Zodiac sign: Taurus"
    elif month == 5 and day >= 21 and day <= 31 or month == 6 and day >= 1 and day <= 20:
        return f"Zodiac sign: Gemini"
    elif month == 6 and day >= 21 and day <= 30 or month == 7 and day >= 1 and day <= 22:
        return f"Zodiac sign: Cancer"
    elif month == 7 and day >= 23 and day <= 31 or month == 8 and day >= 1 and day <= 22:
        return f"Zodiac sign: Leo"
    elif month == 8 and day >= 23 and day <= 31 or month == 9 and day >= 1 and day <= 22:
        return f"Zodiac sign: Virgo"
    elif month == 9 and day >= 23 and day <= 30 or month == 10 and day >= 1 and day <= 22:
        return f"Zodiac sign: Libra"
    elif month == 10 and day >= 23 and day <= 31 or month == 11 and day >= 1 and day <= 21:
        return f"Zodiac sign: Scorpio"
    elif month == 11 and day >= 22 and day <= 30 or month == 12 and day >= 1 and day <= 21:
        return f"Zodiac sign: Sagittarius"
    elif month == 12 and day >= 22 and day <= 31 or month == 1 and day >= 1 and day <= 19:
        return f"Zodiac sign: Capricorn"


def percentage(day, month):
    table = pd.read_csv("data.csv", sep=",")
    maanden = table[table["date_of_month"] == day]
    lijst = maanden[maanden["month"] == month]
    sum_date_births = lijst["births"].sum()
    sum_births = table["births"].sum()
    percentage = (sum_date_births / sum_births) * 100
    rounded = round(percentage, 2)
    return f"Percentage of people born on same day: {rounded} %"


def historical(day, month):
    url = f"https://api.api-ninjas.com/v1/historicalevents?month={month}&day={day}"
    headers = {"X-Api-Key": "bncXEh6rAIsf7zRiKHR3BFJiilzJskTfH5u4Azus"}
    try:
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            data = response.json()
            output = "Historical events:"
            for x in data:
                year = x["year"]
                event = x["event"]
                output += f"\n{year} - {event}"
            return output
    except requests.exceptions.RequestException:
        return "Historical events unavailable (network blocked)"


def one(day, month, year):
    day = int(day)
    month = int(month)
    year = int(year)

    date = dt.datetime(year, month, day).strftime("%d %B %Y")
    dated = date.split()
    month = dated[1].lower()

    url = f"https://www.onthisday.com/date/{year}/{month}/{day}"
    headers = {"User-Agent": "Anthony"}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    evenementen = soup.find_all("li", class_="event")

    for e in evenementen:
        text = e.get_text(" ")
        image = e.find("img")
        if image:
            image.get("src")
            if image.get("alt") == "US":
                output = text[11:].replace("  ", " ")
                return f"Number one hit:{output}"


def wikipedia(day, month):
    date = dt.datetime(2026, month, day)
    dated = date.strftime("%d %B %Y")
    parts = dated.split()
    month = parts[1]

    url = f"https://en.wikipedia.org/wiki/{month}_{day}"
    response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})
    soup = BeautifulSoup(response.text, "html.parser")

    births_header = soup.find("h2", id="Births")
    if births_header is None:
        return "Wikipedia data unavailable"

    entries = []
    current_section = "Unknown"

    for sibling in births_header.parent.find_next_siblings():
        if sibling.get("class") and "mw-heading2" in sibling.get("class", []):
            break
        if sibling.get("class") and "mw-heading3" in sibling.get("class", []):
            current_section = sibling.get_text(strip=True).replace("[edit]", "").strip()
            continue
        for li in sibling.find_all("li"):
            text = li.get_text()
            parts = text.split("–", 1)
            if len(parts) == 2:
                year_str = parts[0].strip().split()[0].strip()
                try:
                    year = int(year_str)
                    entries.append({"section": current_section, "year": year, "text": text})
                except ValueError:
                    pass

    fileName = f"{month}_{day}.txt"
    with open(fileName, 'w', encoding='utf-8') as file:
        current_sec = None
        for e in entries:
            if e["section"] != current_sec:
                current_sec = e["section"]
            outcome = e['text'].replace("died", "d.")
            file.write(f"{outcome}\n")
    return f"File saved as {fileName}"


def age(day, month, year):
    day = int(day)
    month = int(month)
    year = int(year)

    today = dt.datetime.now()
    today_month = today.month
    today_day = today.day
    today_year = today.year

    if today_month < month:
        age = today_year - year - 1
        return f"Today {age} years old!"
    elif today_month == month and today_day < day:
        age = today_year - year - 1
        return f"Today {age} years old!"
    else:
        age = today_year - year
        return f"Today {age} years old!"


def main():
    date = input()
    newDate = parser.parse(date)

    day = newDate.day
    month = newDate.month
    year = newDate.year

    thread_type = input()

    if thread_type == "single":
        print(age(day, month, year))
        print(zodiac(day, month))
        print(percentage(day, month))
        print(historical(day, month))
        print(wikipedia(day, month))
        print(one(day, month, year))

    elif thread_type == "multi":
        results = {}

        def run(func, key, *args):
            results[key] = func(*args)

        threads = [
            threading.Thread(target=run, args=(age,        "age",  day, month, year)),
            threading.Thread(target=run, args=(zodiac,     "zodiac", day, month)),
            threading.Thread(target=run, args=(percentage, "pct",  day, month)),
            threading.Thread(target=run, args=(historical, "hist", day, month)),
            threading.Thread(target=run, args=(wikipedia,  "wiki", day, month)),
            threading.Thread(target=run, args=(one,        "one",  day, month, year)),
        ]

        for t in threads:
            t.start()
        for t in threads:
            t.join()

        for val in results.values():
            print(val)


if __name__ == "__main__":
    main()
```
