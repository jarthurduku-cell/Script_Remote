```python
import json
import requests

city = input("ENTER CITY NAME: ")

def get_weather(city):
    geo_enc_url = f"https://geocoding-api.open-meteo.com/v1/search?name={city}&count=10&language=en&format=json"

    response = requests.get(geo_enc_url)

    print(response)
    print(response.status_code)

    text = response.text
    #print(text)

    data = json.loads(text)
    #print(data)

    info = data["results"][0]
    #print(info)

    for i in data["results"]:
        print(i)
        
    lat = info["latitude"]
    long = info["longitude"]
    print(f"Latitude: {lat}, Longitude: {long}")
    print()

    weather_url = f"https://api.open-meteo.com/v1/forecast?latitude={lat}&longitude={long}&current=temperature_2m,weather_code"
    weather_response = requests.get(weather_url)
    print(weather_response)
    print(weather_response.status_code)

    weather_data = json.loads(weather_response.text)
    print(weather_data)
    print(weather_data["current"])

    current = weather_data["current"]
    weather_code =  current["weather_code"]
    temperature = current["temperature_2m"]
    print(f"Temperature: {temperature}°C")
    print(weather_code)

    if weather_code >= 0 and weather_code <= 3:
        print("clear")
    elif weather_code >= 45 and weather_code <= 48:
        print("fog")
    elif weather_code >= 51 and weather_code <= 57 or weather_code >= 61 and weather_code <= 67 or weather_code >= 80 and weather_code <= 82 or weather_code >= 95 and weather_code <= 99:
        print("rain")
    elif weather_code >= 71 and weather_code <= 77 or weather_code >= 85 and weather_code <= 86:
        print("snow")
    else:
        print("clear")
        
    diction = {"city":city, "condition" : weather_code, "temperature": temperature}
    print(diction)

get_weather(city)


```

```python
import json
import requests 
from ascii_magic import AsciiArt, from_image

def get_pokemon(name): 
     

    name_url = f"https://pokeapi.co/api/v2/pokemon/{name}"
    response = requests.get(name_url)

    pokemon = json.loads(response.text)
    #print(pokemon)

    id = pokemon["id"]


    height = pokemon["height"]
    h_metres = height / 10


    weight = pokemon["weight"]
    w_kg = weight / 10  


    type = []
    #print(pokemon["types"])

    for x in pokemon["types"]:
        type.append(x["type"]["name"])


    stat = {}
    for x in pokemon["stats"]:
        stat[x["stat"]["name"]] = x["base_stat"]


    sprite_url = pokemon["sprites"]["other"]["official-artwork"]["front_default"]


    dict = {"id": id, "height": h_metres, "weight": w_kg, "type": type, "stats": stat, "sprite_url": sprite_url}
    return dict
    
name = input("ENTER POKEMON NAME: ")
print(get_pokemon(name))
pokemon_data = get_pokemon(name)
pic = pokemon_data["sprite_url"]

# download the image
response = requests.get(pic)
with open("pokemon.png", "wb") as f:
    f.write(response.content)

# now convert to ASCII
my_art = from_image("pokemon.png")
my_art.to_terminal()







```

```python
import json
import requests


def get_pokemon(name): 
     

    name_url = f"https://pokeapi.co/api/v2/pokemon/{name}"
    response = requests.get(name_url)

    pokemon = json.loads(response.text)
    #print(pokemon)

    id = pokemon["id"]


    height = pokemon["height"]
    h_metres = height / 10


    weight = pokemon["weight"]
    w_kg = weight / 10  


    type = []
    #print(pokemon["types"])

    for x in pokemon["types"]:
        type.append(x["type"]["name"])


    stat = {}
    for x in pokemon["stats"]:
        stat[x["stat"]["name"]] = x["base_stat"]


    sprite_url = pokemon["sprites"]["other"]["official-artwork"]["front_default"]


    dict = {"id": id, "height": h_metres, "weight": w_kg, "type": type, "stats": stat, "sprite_url": sprite_url}
    return dict

name = input("ENTER POKEMON NAME: ")
# B = 	Buffed types (× 1.5)
# uB =  Debuffed types (× 0.5)
# clear B --> Fire, grass, ground  uB --> ice, water
# rain  B --> water, electric	   uB -->fire, ground
# snow  B --> ice	               uB -->fire, grass, bug
# fog   B --> ghost, dark	       uB --> psychic, fairy
#windy  B --> flying, dragon       uB --> rock, steel

pokemon_data = get_pokemon(name)
print(pokemon_data)
types = pokemon_data["type"]
print(types)
print(len(types))
condition = input("ENTER WEATHER CONDITION: ")
outcome = ""
multiplier = 0

if condition == "clear":
   for x in types:
         if x == "fire" or x == "grass" or x == "ground":
              outcome =  f"{x} buffed in {condition} (x1.5)"
              multiplier = 1.5
              
         elif x == "ice" or x == "water":
              outcome = f"{x} debuffed in {condition} (x0.5)"
              multiplier = 0.5

         else:
              outcome = f"no weather effect in {condition} (×1.0)"
              multiplier = 1.0
    
elif condition == "rain":
    for x in types:
         if x == "water" or x == "electric":
              outcome = f"{x} buffed in {condition} (x1.5)"
              multiplier = 1.5
              
         elif x == "fire" or x == "ground":
              outcome = f"{x} debuffed in {condition} (x0.5)"
              multiplier = 0.5     
         else:
              outcome = f"no weather effect in {condition} (×1.0)"
              multiplier = 1.0
              
elif condition == "snow":
    for x in types:
         if x == "ice":
              outcome = f"{x} buffed in {condition} (x1.5)"
              multiplier = 1.5
              
         elif x == "fire" or x == "grass" or x == "bug":
              outcome = f"{x} debuffed in {condition} (x0.5)"
              multiplier = 0.5
         else:
              outcome = f"no weather effect in {condition} (×1.0)"
              multiplier = 1.0
              
elif condition == "fog":
    for x in types:
         if x == "ghost" or x == "dark":
              outcome = f"{x} buffed in {condition} (x1.5)"
              multiplier = 1.5
                            
         elif x == "psychic" or x == "fairy":
              outcome = f"{x} debuffed in {condition} (x0.5)"
              multiplier = 0.5
         else:
              outcome = f"no weather effect in {condition} (×1.0)"
              multiplier = 1.0
              
elif condition == "windy":
    for x in types:
         if x == "flying" or x == "dragon":
              outcome = f"{x} buffed in {condition} (x1.5)"
              multiplier = 1.5
              
         elif x == "rock" or x == "steel":
              outcome = f"{x} debuffed in {condition} (x0.5)"
              multiplier = 0.5
         else:
              outcome = f"no weather effect in {condition} (×1.0)"
              multiplier = 1.0

print(outcome)
print(float(outcome[-4:-1]))
multiplier = float(outcome[-4:-1])
print(multiplier)

print(pokemon_data["stats"])
for stats in pokemon_data["stats"]:
    pokemon_data["stats"][stats] = int(pokemon_data["stats"][stats] * multiplier)
    
    
print(pokemon_data["stats"])
pokemon_data["outcome"] = outcome
print(pokemon_data)

attack = pokemon_data["stats"]["attack"]
defense = pokemon_data["stats"]["defense"]
damage = int(attack -(defense/2) )


print(f"Damage: {damage}")
```
