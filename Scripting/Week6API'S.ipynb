## JSON AND XML
- JSON is a text based file format mainly used for stroring and transmitting date,
usually this is typically done between a webserver and a client

- XML is more complex, requires a setup of a tag and is worse to process in a lot of coding languages

```python
import json
#string format of json code --> examples
json_data = '{"name": "Anthony", "age" : 30, "study": "computer science"}'
data = json.loads(json_data)#string load
print(data["name"])#keep in mind you're cycling through a dictionary --> result is 

#file load (option 2)
json_file = open("books.json", "r")
data = json.loads(json_file.read())
print()
print(data)
print(data["books"][0]["title"])
## or use this
print(f'First book title: {data["books"][0]["title"]}')

for book in data["books"]:
    print(book["title"])
    
json_file.close()
```

```python
import json, requests 

url = "https://api.kanye.rest"
response = requests.get(url)

#print(response)
#print(response.status_code)

text = response.text
data = json.loads(text)

print(data)
print()
print(f'''Kanyes's daily quote: {data["quote"]}''')
```
